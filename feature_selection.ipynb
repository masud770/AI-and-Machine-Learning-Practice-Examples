```python
import numpy as np
import pandas as pd
```

```python
df= pd.read_csv('/content/HousePrice (1).csv')
df
```

```python
df.isnull().sum()
```

```python
((df.isnull().sum())/len(df))*100
```

```python
df['yr_built']=df['yr_built'].fillna(df['yr_built'].median())
```

```python
df['yr_built'].isnull().sum()
```

```python
df['floors']=df['floors'].fillna(df['floors'].median())
```

```python
((df.isnull().sum())/len(df))*100
```

```python
df =df.dropna()
```

```python
((df.isnull().sum())/len(df))*100
```

```python
df=df.drop('Unnamed: 0',axis =1)
df
```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
```

```python
df['country']=le.fit_transform(df['country'])
df['city'] = le.fit_transform(df['city'])
df['statezip']=le.fit_transform(df['statezip'])
```

```python
df
```

```python
x= df.drop('price',axis=1)
y= df['price']
```

```python
y
```

```python
from sklearn.model_selection  import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
```

```python
fit_features= SelectKBest(score_func=f_classif,k='all')
fit_features.fit(x,y)
```

```python
fs=pd.DataFrame(fit_features.scores_, index=x.columns, columns=['Score Values'])
fs
```

```python
fs=fs.nlargest(8,'Score Values')
```
