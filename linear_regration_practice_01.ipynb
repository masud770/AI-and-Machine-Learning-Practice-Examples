```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("/content/placement.csv")
df
```

```python
 df.isnull().sum()
```

```python
x= df[['placement_exam_marks']]
y= df[['cgpa']]
```

```python
x
```

```python
y
```

```python
plt.scatter(df['placement_exam_marks'],df['cgpa'])
```

```python
plt.scatter(df['placement_exam_marks'],df['cgpa'], marker= '+', color='red')
plt.xlabel('placement_exam_marks')
plt.ylabel('CGPA')
plt.title('Placement')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.25, random_state=1)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
re = LinearRegression()
```

```python
re.fit(x_train,y_train)
```

```python
y_test
```

```python
re.predict(x_test)
```

```python

plt.scatter(df['placement_exam_marks'],df['cgpa'], marker= '+', color='red')
plt.xlabel('placement_exam_marks')
plt.ylabel('CGPA')
plt.title('Placement')
plt.plot(df.placement_exam_marks, re.predict(df[['placement_exam_marks']]))
```
