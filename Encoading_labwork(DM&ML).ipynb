LEBEL ENCODING

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['New District'] = le.fit_transform(df['District'])
df['New Speciality'] = le.fit_transform(df['Speciality'])
df
```

ONE HOT Encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(sparse=False,handle_unknown='error',drop='first')
encoder
```

```python
df
```

```python
df['District'].unique()
```

```python
df['Speciality'].unique()
```

```python
arr = encoder.fit_transform(df[['District','Speciality']])
arr
```

```python
df1=pd.DataFrame(arr)
```

```python
df1
```

```python
df=df.join(df1)
df
```

FEACHER SELECTION

```python
df2= pd.read_csv('/content/heart.csv')
df2
```

```python
x= df2.drop('output',axis=1)
y= df2['output']
df2.corr()
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
```

```python
ft = SelectKBest(score_func= f_classif)
ft.fit(x,y)
```

```python
fs = pd.DataFrame(ft.scores_,index=x.columns, columns=['score values'])
fs
```

```python
des= fs.nlargest(13,'score values')
des
```
